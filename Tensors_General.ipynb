**Create 1D tensor, 2Dtensor and 3D tensors**

```python
import tensorflow as tf
#1D Tensor
x=tf.constant([1,2])
print("1D tensor:")
print(x)
print("Shape of 1D tensor:")
print(x.shape)
#2D Tensor
x1=tf.constant([[1,2],[3,4]])
print("\n")
print("2D tensor:")
print(x1)
print("Shape of 2D tensor:")
print(x1.shape)
#3D Tensor
x2=tf.constant([[[1,2],[3,4]],[[7,8],[9,10]]])
print("\n")
print("3D tensor:")
print(x2)
print("Shape of 3D tensor:")
print(x2.shape)
```

**Create 4-D tensor**

```python
import tensorflow as tf
#3D Tensor
x1=tf.constant([
    [[1,2,3],[4,5,6],[7,8,9]],
    [[1,2,3],[4,5,6],[7,8,9]]
    ])
print("3D Tensor:")
print(x1)
print("3D Tensor Shape:")
print(x1.shape)

#4D Tensor
x2=tf.constant([
    [
    [[1,2],[4,5],[7,8]],
    [[1,2],[4,5],[7,8]],
    ]
    ])
print("4D Tensor:")
print(x2)
print("4D Tensor Shape:")
print(x2.shape)

#5D Tensor
x3=tf.constant([
    [
    [[1,2],[4,5],[7,8]],
    [[1,2],[4,5],[7,8]],
    ]
    ])
print("5D Tensor:")
print(x3)
print("5D Tensor Shape:")
print(x3.shape)
```

Shuffle the tensors in a dataset

```python
import tensorflow as tf

# Create a dataset from a tensor
dataset = tf.data.Dataset.from_tensor_slices([1, 2, 3, 4, 5])

for element in dataset:
  print(element)

# Shuffle the dataset
dataset = dataset.shuffle(buffer_size=3)

# Print the elements of the dataset
print("Shuffled dataset")
for element in dataset:
  print(element)
```
